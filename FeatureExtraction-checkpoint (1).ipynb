```python
import tensorflow
from keras.applications.resnet50 import ResNet50
from keras.applications.vgg16 import preprocess_input
from keras.layers import GlobalMaxPooling2D
import cv2
import numpy as np
from numpy.linalg import norm
```

### ResNet50 pre-trained Deep Learning Model

```python
model=ResNet50(weights='imagenet', include_top=False, input_shape=(224,224,3))
model.trainable=False
```

```python
model=tensorflow.keras.Sequential([
    model, GlobalMaxPooling2D()
])
model.summary()
```

```python
from keras.preprocessing import image
from keras.applications.vgg16 import preprocess_input
import numpy as np

def extract_feature(image_path, model):
    img = image.load_img(image_path, target_size=(224, 224))  # Assuming your model requires 224x224 input size
    img_array = image.img_to_array(img)
    expand_img = np.expand_dims(img_array, axis=0)
    preprocessed_img = preprocess_input(expand_img)
    result = model.predict(preprocessed_img)
    normalized_result = result / np.linalg.norm(result)
    return normalized_result
```

```python
#example
extract_feature("Dataset\9116.jpg",model)
```

```python
import os
from tqdm import tqdm
```

```python
filename=[]
features_list=[]

for file in os.listdir('Dataset'):
    filename.append(os.path.join('Dataset', file))
```

```python
filename[0:5]
```

```python
for file in tqdm(filename):
    features_list.append(extract_feature(file,model))
```

```python
import pickle

pickle.dump(features_list,open('featureVector.pkl','wb'))
pickle.dump(filename,open('filenames.pkl','wb'))
```
